### Merge all the txt files for all years

#### merge all the txt files for ICD10- all years

```python
t0 = time.time()

frames = list()
year = range(1999,2016)
for i in range(0,17):
    filename = str(year[i])+'_tab.txt'
    print(year[i])
    df = pd.read_csv(filename, sep='\t', low_memory=False)
    frames.append(df)
data_all = pd.concat(frames)  
t1 = time.time()

total = t1-t0
print(total)
```

```python
data_all = data_all.reset_index(drop=True)
```

### Rename Sex as 'M' and 'F'

```python
data_all.loc[data_all['Sex']=='1','Sex'] = 'M'
data_all.loc[data_all['Sex']== 1 ,'Sex'] = 'M'
```

```python
data_all.loc[data_all['Sex']=='2','Sex'] = 'F'
data_all.loc[data_all['Sex']== 2, 'Sex'] = 'F'
```

### Education Binning ###

```python
education_value = [-1, 1, 8, 13, 99, 120]
education_groups = ['No formal education', 'elementary school', 'high school','college','Not stated']
```

```python
data_all['Education_ctg'] = pd.cut(data_all['Education'], education_value, labels=education_groups)
```

### Age Binning ###

https://www.cdc.gov/nchs/data/nvsr/nvsr64/nvsr64_02.pdf

```python
age_value = [-1,1,4,14,24,34,44,54,64,74,84,120]
age_groups = ['<1','1-4', '5-14', '15-24', '25-34', '35-44', '45-54', '55-64'
             , '65-74', '75-84', '85-120']
```

```python
data_all['Age_ctg'] = pd.cut(data_all['Age'], age_value, labels=age_groups)
```

### Race Binning ###

```python
Race_value = [-1, 0, 1.5, 2.5, 3.5, 4.5, 5.5, 6.5, 7.5, 19, 29, 39, 49, 59, 69, 79 ]
Race_groups = ['Other races','White', 'Black', 'American Indian ', 'Chinese', 'Japanese', 'Hawaiian',\
                'Filipino','Asian Indian','Korean','Samoan', 'Vietnamese', 'Guamanian',\
                 'Other Asian ','Combined other Asian ']
```

```python
data_all['Race_ctg'] = pd.cut(data_all['Race'], Race_value, labels=Race_groups)
```
